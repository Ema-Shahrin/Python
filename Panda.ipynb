It can be think of as a subset or a structured part of a whole dataset.

 Manipulating a dataset/dataframe


```python
#Removing a row 
dataset.drop(index=0,axis=0)
```

```python
#Removing a col
dataset.drop(index=0,axis=1)
```

```python
# removing a column
diabetes_df.drop(columns='new_BMI',axis=1)
```

```python
# locating a row using index value
dataset.iloc[3]
```

```python
# locating a particular column
print(dataset.iloc[:,0]) # first col
print(dataset.iloc[:,1]) # 2nd col
print(dataset.iloc[:,5]) # 5th col
print(dataset.iloc[:,-1]) # last col
```

Correlation : Finding the relationships between the various columns 
Two types of correlations:
Positive correlation ⇒ one value increases for other value increase 
Negative correlation ⇒ if one decrease for one increase 

Syntax:df.corr()



```python
dataset.corr()
```
